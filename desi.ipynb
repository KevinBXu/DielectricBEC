```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.linspace(0, 10, 1000)

lCDM = lambda z : z ** 0
wgCDM = lambda z : (1+z) ** (3*(1-0.9))
wlCDM = lambda z : (1+z) ** (3*(1-1.1))
```

```python
plt.plot(z, lCDM(z), label=r"$w_{de}=-1$")
plt.ylim([0.4, 2])
plt.xticks([])
plt.yticks([])
plt.xlabel("Time")
plt.ylabel("Energy Density")
plt.legend()
```

```python
plt.figure(figsize=(7, 4))

plt.plot(-z, lCDM(z) / lCDM(10),  label=r"$w=-1$",  color="#4C78A8", linewidth=2.5)
plt.plot(-z, wgCDM(z) / wgCDM(10), label=r"$w>-1$", color="#54A24B", linewidth=2.5)
plt.plot(-z, wlCDM(z) / wlCDM(10), label=r"$w<-1$", color="#E45756", linewidth=2.5)

plt.ylim([0.4, 2.2])
plt.xticks([], fontsize=14)
plt.yticks([], fontsize=14)

plt.xlabel("Time", fontsize=16)
plt.ylabel("Energy Density", fontsize=16)

plt.legend(frameon=False, fontsize=14)
plt.tight_layout()
plt.savefig("wCDM.png")
plt.show()
```
